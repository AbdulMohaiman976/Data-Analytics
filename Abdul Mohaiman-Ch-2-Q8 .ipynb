```python
#d-)
import matplotlib.pyplot as plt

pd.plotting.scatter_matrix(college[['Top10perc','Apps','Enroll']])
plt.show()
```

```python
#e-)
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot without title
sns.boxplot(x='Private', y='Outstate',data=college)

# Axis labels only (optional, but keeps it readable)
plt.xlabel('Private')
plt.ylabel('Outstate')

# Show the plot
plt.show()
```

```python
#f-)
import pandas as pd

# Create new variable 'Elite'
college['Elite'] = pd.cut(college['Top10perc'],
                          bins=[0, 50, 100],  # Use -1 instead of 0 to include 0%
                          labels=['No', 'Yes'])


college['Elite'].value_counts()

sns.boxplot(x='Elite', y='Outstate', data=college)

plt.xlabel('Elite College')
plt.ylabel('Outstate ($)')
plt.show()
```

```python
#g-)

# Create 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Plot histograms in each subplot with different bin sizes
college['Apps'].plot.hist(bins=20, ax=axes[0, 0])
axes[0, 0].set_title('Applications Received')

college['Accept'].plot.hist(bins=30, ax=axes[0, 1])
axes[0, 1].set_title('Students Accepted')

college['Enroll'].plot.hist(bins=15, ax=axes[1, 0])
axes[1, 0].set_title('Students Enrolled')

college['Outstate'].plot.hist(bins=25, ax=axes[1, 1])
axes[1, 1].set_title('Out-of-State Tuition')

# Adjust layout
plt.tight_layout()
plt.show()
```
